```python
con=mysql.connector.connect(user='root',password='123456',host='127.0.0.1',database='w3schools')
cursor=con.cursor()
```

```python
query="""select distinct country from customers"""
cursor.execute(query)
result=cursor.fetchall()
countries=[]
for row in result:
    countries.append(row[0])
countries    
```

```python
query="select count(distinct country) from customers"
cursor.execute(query)
result=cursor.fetchall()
result[0][0],len(countries)
```

```python
query="""select count(*) from customers where country='Mexico' """
cursor.execute(query)
result=cursor.fetchall()
result[0][0]
```

```python
query="select count(*) from customers where country in ('Spain','Germany')"
cursor.execute(query)
result=cursor.fetchall()
result[0][0]
```
